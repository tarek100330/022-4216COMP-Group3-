```python
import pandas as pd 
import matplotlib as plt
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
df = pd.read_csv('finalDataset.csv')
df.head()
```

```python
df.plot.scatter(x = 'Recovered_perDay', y = 'Vaccines_perDay',figsize=(15,6))
plt.title('Vaccines_perDay and Recovered_perDay',fontsize=20,color='Black',loc='Right')
plt.xlabel('Recovered_perDay',color='Red',fontsize=15)
plt.ylabel('Vaccines_perDay',color='Blue',fontsize=15)
plt.savefig('Vaccines_perDay and Recovered_perDay')
```

```python
import pandas as pd
import matplotlib.pyplot as plt



covidDf = pd.read_excel("finalDataset_test.xlsx")

Cumulative_tests = covidDf['Cumulative_recovered_cases']
Cumulative_recovered_cases = covidDf['Cumulative_tests']
plt.figure(figsize=(20,20))
plt.plot(Cumulative_tests, Cumulative_recovered_cases, marker='o')
plt.xlabel('Cumulative_recovered_cases')
plt.ylabel('Cumulative_tests')
plt.ylim(0, 200000)
plt.xlim(0,10000)
plt.title('Relation between Cumulative_recovered_cases and Cumulative_tests ')
plt.show()

```
