```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# set dataset path variable
PATH_DATASET = 'finalDataset.csv'
```

```python
# load dataset into df
df = pd.read_csv(PATH_DATASET, index_col=[0])
# convert df index into proper datetime format
df.index = pd.to_datetime(df.index)

print(df.info())
```

```python
# set figure configurations
plt.figure(figsize=(15,10))
plt.style.use('ggplot')

# plot both columns' df data
plt.plot(df.index, df['Cases_perDay'], label='Cases_perDay')
plt.plot(df.index, df['Deaths_perDay'], label='Deaths_perDay')

# set figure main & axis titles 
plt.suptitle("COVID-19 Cases & Deaths Per Day Relationship", color='black', fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Amount')

# set other figure settings
plt.legend()
plt.tight_layout()

# display plotted figure
plt.show()
plt.clf()
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
ax2 = ax1.twinx()
plt.style.use('ggplot')

# plot both columns' df data
ax1.plot(df.index, df['Cases_perDay'], label='Cases_perDay', color='red')
ax2.plot(df.index, df['Deaths_perDay'], label='Deaths_perDay', color='green')

# set figure main & axis titles 
plt.suptitle("COVID-19 Cases & Deaths Per Day Relationship\nRed Line = Cases Per Day\nGreen Line = Deaths Per Day", 
            color='black', fontweight='bold')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cases_perDay')
ax2.set_ylabel('Deaths_perDay')

# set other figure settings
plt.tight_layout()

# display plotted figure
plt.show()
plt.clf()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15,10))
plt.style.use('ggplot')

# plot both columns' df data
ax[0].plot(df.index, df['Cases_perDay'], label='Cases_perDay', color='red')
ax[1].plot(df.index, df['Deaths_perDay'], label='Deaths_perDay', color='green')

# set figure main & axis titles 
plt.suptitle("COVID-19 Cases & Deaths Per Day Relationship", 
            color='black', fontweight='bold')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Cases_perDay')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Deaths_perDay')

# set other figure settings
ax[0].legend()
ax[1].legend()
plt.tight_layout()

# display plotted figure
plt.show()
plt.clf()
```
